```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statistics import mean
```

```
df_asistencia = pd.read_excel('asistencia_semanal.xlsx')
```

# Informe de Asistencia Semanal por ciclo año 2023
---

## Asistencia Semanal | 1° Ciclo | Marzo 2023

```
# Plot Size
fig = plt.figure(figsize=(10,6))

df_asistencia_1 = df_asistencia[df_asistencia['curso']=='primero básico a']
df_asistencia_2 = df_asistencia[df_asistencia['curso']=='segundo básico a']
df_asistencia_3 = df_asistencia[df_asistencia['curso']=='tercero básico a']
df_asistencia_4 = df_asistencia[df_asistencia['curso']=='cuarto básico a']

x1 = df_asistencia_1['semana']
y1 = df_asistencia_1['porcentaje_asistencia']

x2 = df_asistencia_2['semana']
y2 = df_asistencia_2['porcentaje_asistencia']

x3 = df_asistencia_3['semana']
y3 = df_asistencia_3['porcentaje_asistencia']

x4 = df_asistencia_4['semana']
y4 = df_asistencia_4['porcentaje_asistencia']

# plotting
plt.plot(x1, y1, marker='o', linestyle='-', color='g', label='1° básico')
plt.plot(x2, y2, marker='s', linestyle='--', color='violet', label='2° básico')
plt.plot(x3, y3, marker='p', linestyle='-', color='y', label='3° básico')
plt.plot(x4, y4, marker='D', linestyle='--', color='r', label='4° básico')

for i, v in enumerate(y1):
    if i+1 == 5:
        plt.text(i+1, v-0.8, str(round(v, 1))+'%', verticalalignment='bottom', horizontalalignment='center')
    else:
        plt.text(i+1, v+0.2, str(round(v, 1))+'%', verticalalignment='bottom', horizontalalignment='center')

for i, v in enumerate(y2):
    if i+1 == 3:
        plt.text(i+1, v-0.6, str(round(v, 1))+'%', verticalalignment='baseline', horizontalalignment='center')
    elif i+1 == 4:
        plt.text(i+1, v-0.6, str(round(v, 1))+'%', verticalalignment='baseline', horizontalalignment='center')
    else:
        plt.text(i+1, v+0.2, str(round(v, 1))+'%', verticalalignment='bottom', horizontalalignment='center')

for i, v in enumerate(y3):
    if i+1 != 3:
        plt.text(i+1, v+0.2, str(round(v, 1))+'%', verticalalignment='bottom', horizontalalignment='center')
    else:
        plt.text(i+1, v-0.8, str(round(v, 1))+'%', verticalalignment='bottom', horizontalalignment='center')

for i, v in enumerate(y4):
    if i+1 == 4:
        plt.text(i+1.1, v+0.2, str(round(v, 1))+'%', verticalalignment='baseline', horizontalalignment='center')
    else:
        plt.text(i+1, v+0.2, str(round(v, 1))+'%', verticalalignment='bottom', horizontalalignment='center')

# agregar nombre a los ejes y titulo al grafico
plt.xlabel('N° de Semana')
plt.ylabel('Porcentaje de Asistencia')
plt.title('Asistencia Semanal | 1° Ciclo | Marzo 2023')
plt.legend(bbox_to_anchor=(0.6, 0.21))

plt.xticks([1, 2, 3, 4, 5])

# mostrar plot
plt.show()
```

```
# Plot Size
fig = plt.figure(figsize=(10,6))

df_asistencia_1 = df_asistencia[df_asistencia['curso']=='quinto básico a']
df_asistencia_2 = df_asistencia[df_asistencia['curso']=='sexto básico a']
df_asistencia_3 = df_asistencia[df_asistencia['curso']=='séptimo básico a']
df_asistencia_4 = df_asistencia[df_asistencia['curso']=='octavo básico a']

x1 = df_asistencia_1['semana']
y1 = df_asistencia_1['porcentaje_asistencia']

x2 = df_asistencia_2['semana']
y2 = df_asistencia_2['porcentaje_asistencia']

x3 = df_asistencia_3['semana']
y3 = df_asistencia_3['porcentaje_asistencia']

x4 = df_asistencia_4['semana']
y4 = df_asistencia_4['porcentaje_asistencia']

# plotting
plt.plot(x1, y1, marker='o', linestyle='-', color='lawngreen', label='5° básico')
plt.plot(x2, y2, marker='s', linestyle='--', color='dodgerblue', label='6° básico')
plt.plot(x3, y3, marker='p', linestyle='-', color='y', label='7° básico')
plt.plot(x4, y4, marker='D', linestyle='--', color='tomato', label='8° básico')

for i, v in enumerate(y1):
    if i+1 == 2:
        plt.text(i+1, v-0.7, str(round(v, 1))+'%', verticalalignment='baseline', horizontalalignment='left')
    elif i+1 == 4:
        plt.text(i+1, v-0.7, str(round(v, 1))+'%', verticalalignment='baseline', horizontalalignment='center')
    else:
        plt.text(i+1, v+0.2, str(round(v, 1))+'%', verticalalignment='bottom', horizontalalignment='center')

for i, v in enumerate(y2):
    plt.text(i+1, v+0.1, str(round(v, 1))+'%', verticalalignment='bottom', horizontalalignment='center')

for i, v in enumerate(y3):
    if i+1 != 3:
        plt.text(i+1, v+0.2, str(round(v, 1))+'%', verticalalignment='bottom', horizontalalignment='center')
    else:
        plt.text(i+1, v-0.8, str(round(v, 1))+'%', verticalalignment='bottom', horizontalalignment='center')

for i, v in enumerate(y4):
    if i+1 == 5:
        plt.text(i+1, v-0.8, str(round(v, 1))+'%', verticalalignment='bottom', horizontalalignment='center')
    else:    
        plt.text(i+1, v+0.2, str(round(v, 1))+'%', verticalalignment='bottom', horizontalalignment='center')

# agregar nombre a los ejes y titulo al grafico
plt.xlabel('N° de Semana')
plt.ylabel('Porcentaje de Asistencia')
plt.title('Asistencia Semanal | 2° Ciclo | Marzo 2023')
plt.legend(loc='lower right')

plt.xticks([1, 2, 3, 4, 5])

# mostrar plot
plt.show()
```
